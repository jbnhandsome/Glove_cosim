```python
import torch
import json
import torchtext.vocab as vocab
import random
import time
```

```python
def cos_sim(W, x):
    #W 是总的向量 x是输入单词的向量，k是寻找最接近的k个单词
    W = W.cuda()
    x=x.cuda()
    cos = torch.matmul(W, x.view((-1,))) / (
        (torch.sum(W * W, dim=1) + 1e-9).sqrt() * torch.sum(x * x).sqrt())
    return cos
```

```python
def cos_sort(cos,k):
    _, topk = torch.topk(cos, k,largest=True,sorted=True)
    topk = topk.cpu()
    return topk, [cos[i].item() for i in topk]
```

```python
def find_all(glove,k):
    total_cos={}
    #计算词表中所有的词和其他词的余弦相似度
    count = 0
    start = time.time()
    for i in list(glove.stoi):
        count = count +1
        cos = cos_sim(glove.vectors,glove.vectors[glove.stoi[i]])
        topk,cos = cos_sort(cos,k+1)
        sim=[]
        for j in topk[1:]:
            sim.append(glove.itos[j])
        total_cos[i]=sim
        end = time.time()
        if(count%100 ==0):
            print("现在是",count)
            print(end-start)
    return total_cos
```

```python
use_gpu = torch.cuda.is_available()
cache_dir = ".vector_cache"
glove = vocab.GloVe(name='6B', dim=300, cache=cache_dir)
x = glove
if(use_gpu):
    print("AMD YES!")
    y = find_all(x,100)
else:
    print("lol")
    find_all(x,100)
    with open('file.txt','w') as file:
        file.write(json.dumps(y))
```

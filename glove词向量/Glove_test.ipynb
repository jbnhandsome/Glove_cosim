```python
import torch
import json
import torchtext.vocab as vocab
import random
```

```python
def knn(W, x, k):
    #W 是总的向量 x是输入单词的向量，k是寻找最接近的k个单词
    cos = torch.matmul(W, x.view((-1,))) / (
        (torch.sum(W * W, dim=1) + 1e-9).sqrt() * torch.sum(x * x).sqrt())
    _, topk = torch.topk(cos, k,largest=True,sorted=True)
    topk = topk.cpu().numpy()
    return topk, [cos[i].item() for i in topk]
```

```python
def search_sim_token(query_token, k, glove):
    #第一个参数 查询的 单词 第二个参数 knn 第三个参数 与训练好的词向量
    topk, cos = knn(glove.vectors,
                    glove.vectors[glove.stoi[query_token]], k+1)
    for i, c in zip(topk[1:], cos[1:]):  # 除去输入词
        print('余弦相似度=%.3f: %s' % (c, (glove.itos[i])))
```

```python
def find(glove,k):
    total_cos={}
    #计算词表中所有的词和其他词的余弦相似度
    # for i in random.sample(list(glove.stoi),20):
    #从词表中随机抽取20个单词，计算他们和剩余单词的余弦相似度
    for i in random.sample(list(glove.stoi),20):
        topk, cos = knn(glove.vectors,
                    glove.vectors[glove.stoi[i]], k+1)
        cos=[]
        for j in topk[1:]:
            cos.append(glove.itos[j])
            # print(glove.itos[j])
        total_cos[i]=cos
        # with open('file.txt','w') as file:
        #     file.write(json.dumps(total_cos))
    return total_cos
        
        
```

```python
cache_dir = ".vector_cache"
glove = vocab.GloVe(name='6B', dim=300, cache=cache_dir)
print("一共包含%d个词。" % len(glove.stoi))
total_cos=find(glove,100)
with open('file.txt','w') as file:
    file.write(json.dumps(total_cos))
#随机查询一个单词的余弦相似度
search_sim_token('ice',100,glove)
```
